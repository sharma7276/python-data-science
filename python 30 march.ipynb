# Joins

```python
#pandas package used for dataframes
#inner join:- only common record will participate (intersection )
#left outer join:- all the record of left table and intersection.
#right outer join:- all record from right table and intersection
#full outer join:- all the record from both the table
```

```python
import pandas as pd
```

```python
student_details = { 'stu_id' : [1,2,3,4,5],
                    'stu_name' : ['a','b','c','d','e'],
                    'Region' : ['n','s','e','w','w']}
```

```python
sales_details = {'st_id' : [1,2,3,4,6],
                 'sub':['py','r','tab','sql','py'],
                 'sales':[250,344,434,343,343] }
```

```python
student_details = pd.DataFrame(student_details)
```

```python
sales_details = pd.DataFrame(sales_details)
```

```python
student_details
```

```python
sales_details
```

```python
#to apply join merge() is used
#since no common:- left_on='stu_id',right_on='st_id'
```

```python
#left outer join
pd.merge(student_details,sales_details, left_on='stu_id',right_on='st_id',how='left' )
```

```python
#right outer join
pd.merge(student_details,sales_details, left_on='stu_id',right_on='st_id',how='right' )
```

```python
#full outer join
pd.merge(student_details,sales_details, left_on='stu_id',right_on='st_id',how='outer' )
```

```python
sales_details = {'stu_id' : [1,2,3,4,6],
                 'sub':['py','r','tab','sql','py'],
                 'sales':[250,344,434,343,343] }
```

```python
#when common column we use ON
pd.merge(student_details,sales_details, on = 'stu_id',how='right' )
```

```python
df2=pd.DataFrame(np.array([['a',5,19],['b',4,16],['c',2,9]]), columns=['name','attr1','attr2'])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
#first merger df1 and df2 then merge output table with df3
```

```python
pd.merge(pd.merge(df1,df2,on='name'),df3,on='name')
```

```python
#other way to join multiple table
df1.merge(df2, on='name').merge(df3,on='name')
```
